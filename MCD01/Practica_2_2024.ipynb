15. Escriba una función que devuelve el $i-esimo$ número de la sucesión de Fibonacci  (https://en.wikipedia.org/wiki/Fibonacci_number)

```python
def fibo(i):
    inicio = [0, 1]
    while len(inicio) <= i:
        siguiente = inicio[-1] + inicio[-2]
        inicio.append(siguiente)
    return inicio[i]

fibo(6)
```

terminado en 00), en cuyo caso deberá ser divisible por 400. Así, por ejemplo, 1800 no es bisiesto,

```python
def es_bisiesto(n):
    if  (str(n)[-2:]== "00" and n%400==0) :
        return('bisiesto')
    elif (str(n)[-2:]!= "00" and n%4==0): 
        return(' es bisietso')
    else: return('no es bisiesto')
es_bisiesto(1800)
    
```

los números de l que son divisibles por n.

```python
def func(l,n):
    result=[]
    for item in l:
        if item % n==0:
            result.append(item)
    return(result)
func([1,2,4,10,17],2)
```
